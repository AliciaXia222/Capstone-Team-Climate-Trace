# Miami EUI

```python
import pandas as pd
import numpy as np
#The data comes from https://www.benchmark305.com/map
# --> Download 2023 Dataset

eui_miami_path = "https://docs.google.com/spreadsheets/d/1ctfi0LxU9b35pIiZ7pgN_iCNm8zLuJdF/export?format=csv"

eui_miami_df = pd.read_csv(eui_miami_path, skiprows=3)

latitude = 25.7617
longitude = -80.1918

eui_miami_df['latitude'] = latitude
eui_miami_df['longitude'] = longitude
```

```python
property_type_classification = {
    # Residential
    "Multifamily Housing": "Residential",

    # Non-Residential
    "Casino": "Non-Residential",
    "Financial Office": "Non-Residential",
    "Hotel": "Non-Residential",
    "Lifestyle Center": "Non-Residential",
    "Medical Office": "Non-Residential",
    "Mixed Use Property": "Non-Residential",
    "Museum": "Non-Residential",
    "Office": "Non-Residential",
    "Other": "Non-Residential",
    "Other - Education": "Non-Residential",
    "Other - Lodging/Residential": "Non-Residential",
    "Other - Mall": "Non-Residential",
    "Other - Recreation": "Non-Residential",
    "Parking": "Non-Residential",
    "Performing Arts": "Non-Residential",
    "Police Station": "Non-Residential",
    "Refrigerated Warehouse": "Non-Residential",
    "Residential Care Facility": "Non-Residential",
    "Self-Storage Facility": "Non-Residential"
}


eui_miami_df["Property Type Category"] = eui_miami_df["Primary Property Type**"].map(property_type_classification)

```

```python
kBtu_to_kWh = 0.2931
ft2_to_m2 = 0.092903

eui_miami_df = eui_miami_df[eui_miami_df["Weather Normalized Site EUI**"] > 0]

eui_miami_df['EUI (kWh/m2/year)'] = (eui_miami_df['Weather Normalized Site EUI**']) * kBtu_to_kWh / ft2_to_m2

eui_miami_df["Residential EUI (kWh/m2/year)"] = eui_miami_df.apply(
    lambda x: x["EUI (kWh/m2/year)"] if x["Property Type Category"] == "Residential" else None, axis=1
)

eui_miami_df["Non-residential EUI (kWh/m2/year)"] = eui_miami_df.apply(
    lambda x: x["EUI (kWh/m2/year)"] if x["Property Type Category"] == "Non-Residential" else None, axis=1
)

eui_miami_df["Residential Area (ft²)"] = eui_miami_df.apply(
    lambda x: x[" Assessor Gross Floor Area* "] if x["Property Type Category"] == "Residential" else None, axis=1
)

eui_miami_df["Non-residential Area (ft²)"] = eui_miami_df.apply(
    lambda x: x[" Assessor Gross Floor Area* "] if x["Property Type Category"] == "Non-Residential" else None, axis=1
)

eui_miami_df
```

```python
eui_miami_df['Residential Area (ft²)'] = eui_miami_df['Residential Area (ft²)'].replace('None', np.nan)
eui_miami_df['Non-residential Area (ft²)'] = eui_miami_df['Non-residential Area (ft²)'].replace('None', np.nan)

eui_miami_df['Residential Area (ft²)'] = eui_miami_df['Residential Area (ft²)'].astype(str).str.replace(',', '')
eui_miami_df['Non-residential Area (ft²)'] = eui_miami_df['Non-residential Area (ft²)'].astype(str).str.replace(',', '')

eui_miami_df['Residential Area (ft²)'] = pd.to_numeric(eui_miami_df['Residential Area (ft²)'], errors='coerce')
eui_miami_df['Non-residential Area (ft²)'] = pd.to_numeric(eui_miami_df['Non-residential Area (ft²)'], errors='coerce')
```

```python
eui_miami_df = eui_miami_df[[ 
    'latitude', 'longitude', 
    'Residential EUI (kWh/m2/year)', 'Non-residential EUI (kWh/m2/year)', 
    'Residential Area (ft²)',  'Non-residential Area (ft²)'
]]

eui_miami_df.head()
```

```python
eui_miami_df.to_csv("../data/02_interim/eui_miami.csv", index=False)
```
