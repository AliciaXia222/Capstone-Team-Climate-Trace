```python
#1 kBtu = 0.2931 kWh
#1 ft² = 0.092903 m²

# Conversion factors
kBtu_to_kWh = 0.2931
ft2_to_m2 = 0.092903

eui_california_df = eui_california_df[eui_california_df["Weather Normalized Site EUI (kBtu/ft²)"] > 0]


eui_california_df['EUI (kWh/m2/year)'] = (eui_california_df['Weather Normalized Site EUI (kBtu/ft²)'])* kBtu_to_kWh  / ft2_to_m2


eui_california_df["Residential EUI (kWh/m2/year)"] = eui_california_df.apply(
    lambda x: x["EUI (kWh/m2/year)"] if x["Property Type Category"] == "Residential" else None, axis=1
)

eui_california_df["Non-residential EUI (kWh/m2/year)"] = eui_california_df.apply(
    lambda x: x["EUI (kWh/m2/year)"] if x["Property Type Category"] == "Non-Residential" else None, axis=1
)

##
eui_california_df["Residential Area (ft²)"] = eui_california_df.apply(
    lambda x: x["Property GFA - Calculated (Buildings) (ft²)"] if x["Property Type Category"] == "Residential" else None, axis=1
)

eui_california_df["Non-residential Area (ft²)"] = eui_california_df.apply(
    lambda x: x["Property GFA - Calculated (Buildings) (ft²)"] if x["Property Type Category"] == "Non-Residential" else None, axis=1
)
```

```python
eui_california_df = eui_california_df[[ 
    'latitude', 'longitude', 
    'Residential EUI (kWh/m2/year)', 'Non-residential EUI (kWh/m2/year)', 
    'Residential Area (ft²)',  'Non-residential Area (ft²)'
]]

eui_california_df.head()
```
