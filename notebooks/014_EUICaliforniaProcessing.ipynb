```python
#1 kBtu = 0.2931 kWh
#1 ft² = 0.092903 m²

direct_energy_columns = [
    'Natural Gas Use (kBtu)', 
    'Fuel Oil #2 Use (kBtu)', 
    'Diesel Use (kBtu)', 
    'Propane Use (kBtu)', 
    'District Steam Use (kBtu)', 
    'District Hot Water Use (kBtu)', 
    'District Chilled Water Use (kBtu)'
]

indirect_energy_columns = [
    'Electricity Use - Grid Purchase (kBtu)', 
    'Electricity Use – Generated from Onsite Renewable Systems and Used Onsite (kBtu)'
]

total_energy_columns = direct_energy_columns + indirect_energy_columns

eui_california_df['Direct Energy Use (kBtu)'] = eui_california_df[direct_energy_columns].sum(axis=1)

eui_california_df['Indirect Energy Use (kBtu)'] = eui_california_df[indirect_energy_columns].sum(axis=1)


eui_california_df['Total Energy Use (kBtu)'] = eui_california_df[total_energy_columns].sum(axis=1)

#This aprox is comparable with Weather Normalized Site EUI (kBtu/ft²)	
#eui_california_df['Total EUI (kBtu/ft²)'] = eui_california_df['Total Energy Use (kBtu)'] / eui_california_df['Property GFA - Calculated (Buildings) (ft²)']

eui_california_df = eui_california_df[eui_california_df["Direct Energy Use (kBtu)"] != 0]

# Conversion factors
kBtu_to_kWh = 0.2931
ft2_to_m2 = 0.092903

eui_california_df['Direct EUI (kWh/m²)'] = (
    eui_california_df['Direct Energy Use (kBtu)'] * kBtu_to_kWh
) / (eui_california_df['Property GFA - Calculated (Buildings) (ft²)'] * ft2_to_m2)


eui_california_df["Residential EUI (kWh/m²/year)"] = eui_california_df.apply(
    lambda x: x["Direct EUI (kWh/m²)"] if x["Property Type Category"] == "Residential" else None, axis=1
)

eui_california_df["Non-residential EUI (kWh/m²/year)"] = eui_california_df.apply(
    lambda x: x["Direct EUI (kWh/m²)"] if x["Property Type Category"] == "Non-Residential" else None, axis=1
)
```
