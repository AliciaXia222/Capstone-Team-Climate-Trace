```python
import pandas as pd
# https://energy.cusp.nyu.edu/#/

nyc_data_path = "https://drive.google.com/uc?id=1Sn4rBiOWyfxrLJCx5YnMosXMRe6wjLSb&export=download"

nyc_data_df = pd.read_csv(nyc_data_path)

nyc_data_df.head()
```

```python
import os
import gdown

def download_file_from_drive(file_url, file_name, output_dir="../../drive_downloaded_files"):
    """
    Downloads a file from Google Drive.

    Parameters:
        file_url (str): The Google Drive URL to download the file from.
        file_name (str): The name to save the downloaded file as.
        output_dir (str): The directory where the downloaded file will be saved. Defaults to "../../drive_downloaded_files".
    """
    # Create the full output path
    output_path = os.path.join(output_dir, file_name)

    # Create the folder if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Download the file if it doesn't exist
    if not os.path.exists(output_path):
        print(f"Downloading the file from {file_url}...")
        gdown.download(file_url, output_path, quiet=False)
    else:
        print(f"The file {output_path} already exists.")

    return output_path  # Return the path of the downloaded file


# Google Drive file URL and name for pluto_24v4_1
drive_url_pluto = "https://drive.google.com/uc?export=download&id=1z6x82N4_AxdInTATbp26lqRnuBNknLRy"
file_name_pluto = "pluto_24v4_1.csv"

# Download the file
pluto_file_path = download_file_from_drive(drive_url_pluto, file_name_pluto)

# Check if the file was downloaded successfully
print(f"File downloaded and saved at: {pluto_file_path}")
```

```python
pluto_df = pd.read_csv(pluto_file_path)
pluto_df = pluto_df[['bbl', 'latitude', 'longitude']]
pluto_df.head()
```

```python
kBtu_to_kWh = 0.2931
ft2_to_m2 = 0.092903

nyc_data_df = nyc_data_df[nyc_data_df["eui"] > 0]

nyc_data_df['EUI (kWh/m2/year)'] = (nyc_data_df['eui']) * kBtu_to_kWh / ft2_to_m2

nyc_data_df["Residential EUI (kWh/m2/year)"] = nyc_data_df.apply(
    lambda x: x["EUI (kWh/m2/year)"] if x["Property Type Category"] == "Residential" else None, axis=1
)

nyc_data_df["Non-residential EUI (kWh/m2/year)"] = nyc_data_df.apply(
    lambda x: x["EUI (kWh/m2/year)"] if x["Property Type Category"] == "Non-Residential" else None, axis=1
)

nyc_data_df["Residential Area (ft²)"] = nyc_data_df.apply(
    lambda x: x["gfa"] if x["Property Type Category"] == "Residential" else None, axis=1
)

nyc_data_df["Non-residential Area (ft²)"] = nyc_data_df.apply(
    lambda x: x["gfa"] if x["Property Type Category"] == "Non-Residential" else None, axis=1
)
```

```python
nyc_data_df = nyc_data_df[['latitude', 'longitude', 
                           'Residential EUI (kWh/m2/year)', 
                           'Non-residential EUI (kWh/m2/year)',
                           'Residential Area (ft²)', 
                           'Non-residential Area (ft²)']]
```
